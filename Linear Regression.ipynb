```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
```

```python
file = r"C:\Users\abhis\ML Lab\Linear Regression\Regression-AreaPrice (1) - Regression-AreaPrice (1).csv"
df = pd.read_csv(file)
print (df)
```

```python
print('Defining X & y variable')
X = df['Area']
y = df['Price']
plt.scatter(X, y ,color = 'blue')
```

```python
print('Splitting the data into Train & Test ')
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
plt.scatter(X_train, y_train, color = 'blue')
```

```python
print('Training the model using X_train, y_train')
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1, 1), y_train)
```

```python
print('Predicting using the trained model X_test')
y_pred = lr.predict(X_test.values.reshape(-1, 1))
print(y_test)
print(y_pred)
```

```python
plt.scatter(X_train, y_train, color='blue')
plt.scatter(X_test, y_test, color = 'red')
plt.xticks()
plt.yticks()
plt.show()
```

```python
print("Finding intercept & coeff")
print('Intercept', lr.intercept_)
print('coefficent', lr.coef_)
print(lr.coef_, 'Xt', lr.intercept_)
```
